**Importanto a Biblioteca Pandas** e lendo o arquivo **Base_despadronizada**

```python
import pandas as pd

dados_alunos = pd.read_excel("Base_despadronizada.xlsx")
```

**Padronizar a coluna Sexo:**

```python
def padronizacao_genero(genero):
    genero = genero.lower().strip() #Converte para minúsculo e sem espaços
    if genero == 'Masculino' or genero == 'Feminino':
        return genero
    else:
        if genero and genero[0] == 'm':
            return 'Masculino'
        elif genero and genero[0] == 'f':
            return 'Feminino'

dados_alunos['sexo'] = dados_alunos['sexo'].apply(padronizacao_genero)
display(dados_alunos)
```

**Padronizar as colunas de Notas:**

Matemática:

```python
for idx, nota in dados_alunos["nota_matematica"].items():
    s = str(nota)
    if ":" in s:
        data_str = s.split(" ")[0]
        partes = data_str.split("-")
        if len(partes) == 3:
            ano, mes, dia = partes
            novo_valor = f"{int(dia)},{int(mes)}"
        else:
            novo_valor = s
    elif "." in s:
        novo_valor = s.replace(".", ",")
    else:
        novo_valor = f"{s},0"
    # Atualiza o valor no DataFrame
    dados_alunos.at[idx, "nota_matematica"] = novo_valor

display(dados_alunos)
```

Português:

```python
for idx, nota in dados_alunos["nota_portugues"].items():
    s = str(nota)
    if ":" in s:
        data_str = s.split(" ")[0]
        partes = data_str.split("-")
        if len(partes) == 3:
            ano, mes, dia = partes
            novo_valor = f"{int(dia)},{int(mes)}"
        else:
            novo_valor = s
    elif "." in s:
        novo_valor = s.replace(".", ",")
    else:
        novo_valor = f"{s},0"
    # Atualizando o valor no DataFrame
    dados_alunos.at[idx, "nota_portugues"] = novo_valor

display(dados_alunos)
```

Frequência:

```python
dados_alunos["frequencia"] = dados_alunos["frequencia"].astype(str).str.replace(".", ",")
display(dados_alunos)
```

**Criar nova coluna de Media e Aprovado:**

```python
dados_alunos['media'] = ""
dados_alunos['aprovado'] = ""

for idx in range(len(dados_alunos)):
    nota_matematica = dados_alunos.loc[idx, 'nota_matematica']
    nota_portugues = dados_alunos.loc[idx, 'nota_portugues']
    frequencia = dados_alunos.loc[idx, 'frequencia']
    
    nota_matematica = float(nota_matematica.replace(",", "."))
    nota_portugues = float(nota_portugues.replace(",", "."))
    frequencia = float(frequencia.replace(",", "."))
    
    # Calcula a média
    media = (nota_matematica + nota_portugues + (frequencia / 10)) / 3

    # Define se o aluno foi aprovado (média >= 7)
    if media >= 7:
        dados_alunos.at[idx, "aprovado"] = 'Sim'
    else:
        dados_alunos.at[idx, "aprovado"] = 'Não'

    media_formatada = f"{media:.1f}".replace(".", ",")
    
    dados_alunos.at[idx, "media"] = media_formatada

display(dados_alunos)
```

**Transformando o DataFrame em um arquivo .xlsx**

```python
dados_alunos.to_excel("Base_padronizada.xlsx", index=False)
```
